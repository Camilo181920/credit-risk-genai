# Feature Engineering

En esta etapa se construyen nuevas variables que faciliten la identificación del riesgo crediticio.

```python
import streamlit as st

from src.ui.forms import customer_form
from src.services.credit_service import CreditService


st.title("🔍 Credit Assessment")


customer = customer_form()


if st.button("Evaluate Customer"):

    service = CreditService()

    assessment = service.evaluate(
        customer
    )


    col1, col2 = st.columns(2)


    col1.metric(
        "Probability of Default",
        f"{assessment.probability:.2%}"
    )


    col2.metric(
        "Risk Level",
        assessment.risk_level
    )


    st.divider()


    st.subheader(
        "Main Risk Factors"
    )


    for feature in assessment.top_features:

        st.write(
            f"- {feature}"
        )
```
