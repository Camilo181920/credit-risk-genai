# Model Training

Comparación de diferentes algoritmos de clasificación para estimar riesgo crediticio.

```python
from src.data import load_data

from src.feature_engineering import create_customer_profile

df = load_data("german_credit.csv")

df = create_customer_profile(df)
```

```python
from src.preprocessing import (
    split_features_target,
    split_train_test
)

X, y = split_features_target(df)

X_train, X_test, y_train, y_test = split_train_test(
    X,
    y
)
```

```python
from src.model_utils import identify_columns

numeric, categorical = identify_columns(X)
```

```python
from src.train import (
    available_models,
    train_pipeline
)

from src.evaluate import evaluate_model

results = {}

for name, model in available_models().items():

    pipeline = train_pipeline(

        model,

        numeric,

        categorical,

        X_train,

        y_train

    )

    results[name] = evaluate_model(

        pipeline,

        X_test,

        y_test

    )
```

```python
import pandas as pd

results = pd.DataFrame(results).T

results
```

```python
from src.save_model import save_model

best_model = pipeline

save_model(
    best_model,
    "models/credit_model.pkl"
)
```
